```python
import pandas as pd
import numpy as np
```

*Create header column for Data Frame*

```python
df = pd.DataFrame(columns=['PostalCode','Borough','Neighborhood']) 
df
cols =df.columns
cols
```

*Time for Soup*

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

*Confirms page exists*

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
response
```

*Grabs table with Postal Codes in Toronto* 

```python
soup = BeautifulSoup(response.text, 'html.parser')
tbl = soup.find_all('table')[0]
tbl
```

*Filters values and puts them into their entries*

```python
j=0
for tr_tag in tbl.find_all('tr'):
    i=0
    j+=1
    for td_tag in tr_tag.find_all('td'):
        df.loc[j,cols[i]]=td_tag.string
        i+=1
df
```

```python
df.drop_duplicates(subset=['PostalCode','Neighborhood'])  
df = df[df.Borough != 'Not assigned\n']
df
```
